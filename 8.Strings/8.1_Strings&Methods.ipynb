```python
a,b=100,10
print("{} + {} = {}".format(a,b,a+b))
print("{} - {} = {}".format(a,b,a-b))
print("{} * {} = {}".format(a,b,a*b))
print("{} / {} = {}".format(a,b,a/b))
print("{} % {} = {}".format(a,b,a%b))
print("{} // {} = {}".format(a,b,a//b))
```

```python
a=17500
b=10
print(f"{a} + {b}= {a+b}")
print(f"{a} - {b}= {a-b}")
print(f"{a} * {b}= {a*b}")
print(f"{a} / {b}= {a/b}")
print(f"{a} % {b}= {a%b}")
print(f"{a} // {b}= {a//b}")
```

```python
a = "           Good morning       "
print("Before  operation = ",a,end="\n\n") 

a = a.lstrip() # trailing white space will be removed from left.
print("After operation = ",a) 
```
