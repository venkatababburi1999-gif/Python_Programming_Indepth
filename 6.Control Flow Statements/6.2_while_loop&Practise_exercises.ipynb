```python

num2 = str(5082)

smallest = int(num2[0])
largest = int(num2[0])

for i in num2:
    digit_int = int(i)

    if digit_int > largest:
        largest = digit_int

    if digit_int < smallest:
        smallest = digit_int

print("Smallest:", smallest)
print("Largest:", largest)
```
