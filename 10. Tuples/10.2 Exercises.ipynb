```python
#Compare two tuples and find out which one is greater and why?
t1 = (1, 2, 3)
t2 = (1, 2, 4)
print(f"Tuple1: {t1}")
print(f"Tuple2: {t2}")
if t1>t2:
    print(f"Tuple1: {t1} is greater than {t2}")
elif t1<t2:
    print(f"Tuple1: {t1} is less than {t2}")
else:
    print(f"Both tuples are equal")
```

```python
#Write a code to filter out students with scores less than 90 from a given a list of tuples.
students = [('Alice', 85), ('Bob', 92), ('Charlie', 78)]
f=[item for item in students if item[1]>90]
print(f)
```

```python
students = [('Alice', 85), ('Bob', 92), ('Charlie', 78), ('David', 95)]
print(f"Original student list: {students}")

high_achievers_loop = []
for student in students:
  if student[1] >= 90:
    high_achievers_loop.append(student)
print(f"Students with scores 90 or above (loop method): {high_achievers_loop}")
```

```python
#Given a tuple of numbers, create a new tuple where each number is squared
t = (1, 2, 3, 4)
t2=tuple(map(lambda x:x**2,t))
print(t2)
```

```python
t = (1, 2, 3, 4)
l=list(t)
print(f"The original tuple:{t} and converted list:{l}")
empty_list=[]
for i in l:
    empty_list.append(i**2)
squared_tuple=tuple(empty_list)
print(f"The original tuple :{t} and squared tuple is:{squared_tuple}")
```

```python
# 1. Sort a tuple by 2nd item
tuple1 = (('a', 23), ('b', 37), ('c', 11), ('d', 29))
print(f"Original tuple: {tuple1}")
list1=list(tuple1)
print(list1)
sorted_list=sorted(list1,key=lambda item:item[1])
print(sorted_list)
```

```python
tuple1 = (45, 45, 45, 45)
def check(t):
    Flag=True
    if any(i!=t[0] for i in t):
        Flag=False
        return Flag
    else:
        return Flag
a=check(tuple1)
print(a)
```
