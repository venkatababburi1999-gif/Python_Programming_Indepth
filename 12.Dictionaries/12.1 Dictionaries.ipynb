# Characteristics of Dictionaries in Python

Dictionaries in Python are a built-in data type used to store data in the form of **keyvalue pairs**. They are optimized for fast lookups, insertion, and deletion.

---

## 1. Ordering

### Python 3.6 and Lower Versions
- Dictionaries are **unordered** (officially).
- Items do not have a fixed sequence.
- There is **no index-based access**.
- Keys act as logical identifiers, not positional indexes.
- The iteration order is **not guaranteed**.

### Python 3.7 and Higher Versions
- Dictionaries are **ordered**.
- They preserve the **insertion order**.
- Once a key is inserted, its position remains consistent unless modified.
- This behavior is part of the **Python language specification**.

> In CPython 3.6, insertion order was preserved as an implementation detail,  
> but from Python 3.7 onward, it is officially guaranteed.

---

## 2. KeyValue Pair Structure
- Data is stored as **key : value** pairs.
- Keys act as **logical indexes**.
- Values can be of any data type.
- Internally implemented using a **Hash Table**.
- A **hash function** is applied to keys to determine storage locations.

Example:
```python
employee = {
    "id": 101,
    "name": "John",
    "role": "Developer"
}


# Python Dictionary Notes

## 3. Uniqueness of Keys
- Every key in a dictionary must be **unique**.
- Duplicate keys are **not allowed**.
- If a key is repeated, the **most recent value replaces the old one**.

```python
data = {"x": 10, "x": 20}
print(data)   # {'x': 20}
```
## 4. Mutability
- Dictionaries are **mutable**.
- Elements can be modified after creation.

### Supported Operations
- Add new keyvalue pairs  
- Update existing values  
- Remove keyvalue pairs
## 5. Key Restrictions

- Keys must be of **immutable** data types.
- Mutable objects **cannot** be used as dictionary keys.

### Valid Key Types
- int  
- float  
- str  
- tuple  

### Invalid Key Types
- list  
- set  
- dict
## Stores data as keyvalue pairs

- Ordered from Python 3.7 onwards

- Keys are unique

- Mutable collection

- Implemented using a hash table

- Optimized for fast access

````python
# address dictionary to store person address
address = {"state": "Texas", 'city': 'Houston'}

# dictionary to store person details with address as a nested dictionary
person = {'name': 'Jessa', 'company': 'Google', 'address': address}
# Display dictionary
print("person:", person)
````

````python
# each dictionary will store data of a single student
jessa = {'name': 'Jessa', 'state': 'Texas', 'city': 'Houston', 'marks': 75}
emma = {'name': 'Emma', 'state': 'Texas', 'city': 'Dallas', 'marks': 60}
kelly = {'name': 'Kelly', 'state': 'Texas', 'city': 'Austin', 'marks': 85}
mul={'jessa':jessa,'emma':emma,'kelly':kelly}
print(mul)
````

````python
for k,v in mul.items():
    for k1,v1 in v.items():
        print(k1,":",v1,end=" ")
    print()
````

````python
dict1 = {'c': 45, 'b': 95, 'a': 35}

# sorting dictionary by keys
print(sorted(dict1.items()))
# Output [('a', 35), ('b', 95), ('c', 45)]

# sort dict eys
print(sorted(dict1))
# output ['a', 'b', 'c']

# sort dictionary values
print(sorted(dict1.values()))
# output [35, 45, 95]
````
