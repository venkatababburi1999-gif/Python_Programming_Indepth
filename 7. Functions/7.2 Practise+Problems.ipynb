```python
#Write a program to create a recursive function that calculates the sum of numbers from 0 to 10.

#A recursive function is a function that calls itself repeatedly.
def rec(n):
    if n:
        n=n+rec(n-1)
        return n
    else:
        return 0
res=rec(10)
print(res)
```
