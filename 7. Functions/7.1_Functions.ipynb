# Python Functions

1. In Python, the function is a block of code defined with a name. We use functions whenever we need to perform the same task multiple times without writing the same code again. It can take arguments and returns the value.

2. Python has a DRY principle like other programming languages. DRY stands for Donâ€™t Repeat Yourself. Consider a scenario where we need to do some action/task many times. We can define that action only once using a function and call that function whenever required to do the same activity.

3. Function improves efficiency and reduces errors because of the reusability of a code. Once we create a function, we can call it anywhere and anytime. The benefit of using a function is reusability and modularity.

Creating a Function in Python
Use the following steps to define a function in Python.
Steps to Define a Function
- Use the def keyword followed by the function name to define a function.
- Pass the required number of parameters inside parentheses. (Parameters are optional.)
- Write the function body as an indented block of code that performs the desired action.
- Python does not use curly braces for function bodies.
- Proper indentation is essential to separate code blocks.
- Incorrect indentation will result in an error.

# Python Function Arguments
The argument is a value, a variable, or an object that we pass to a function or method call. In Python, there are four types of arguments allowed.
- Positional arguments
- keyword arguments
- Default arguments
- Variable-length argument
  
**Note: A function must be called with the correct number of arguments by default. For example, the above function expects 3 arguments, so you have to call the my_sum() function with 3 arguments; otherwise, you will get an error.**


## Default Arguments
> In a function, arguments can have default values. We assign default values to the argument using the â€˜=â€™ (assignment) operator at the time of function definition. You can define a function with any number of default arguments.

> The default value of an argument will be used inside a function if we do not pass a value to that argument at the time of the function call. Due to this, the default arguments become optional during the function call.

> It overrides the default value if we provide a value to the default arguments during function calls.
---

````python
def add(a,b,c=10):
    print("The value of a is:",a)
    print("The value of b is:",b)
    print("The value of c is:",c)
    return a+b+c
addition=add(10,20)
print(addition)
````

````python
add(10,20,30)
````

## Keyword Arguments
- Usually, at the time of the function call, values get assigned to the arguments according to their position. So we must pass values in the same - sequence defined in a function definition.  
- For example, when we call student('Jon', 12, 'Five', 'ABC School'), the value â€œJonâ€ gets assigned to the argument name, and similarly, 12 to age and so on as per the sequence.  

- We can alter this behavior using a keyword argument.

- Keyword arguments are those arguments where values get assigned to the arguments by their keyword (name) when the function is called. It is preceded by the variable name and an (=) assignment operator. The Keyword Argument is also called a named argument.

We saw how to use *args. Now letâ€™s see how to use the **kwargs argument. The **kwargs allow you to pass multiple keyword arguments to a function. Use the **kwargs if you want to handle named arguments in a function.

Use the unpacking operator(**) to define variable-length keyword arguments. Keyword arguments passed to a kwargs are accessed using key-value pair (same as accessing a dictionary in Python).

````python
def subjects(**marks):
    sum=0
    print(marks)
    for i in marks:
        sub_name=i
        sub_marks=marks[sub_name]
        print(sub_name,sub_marks)
````

````python
subjects(english=100,maths=99,science=85)
````

````python
#recursive functions
def fact(n):
    if n==0:
        return 1
    else:
        result=n*fact(n-1)
    return result
fact(5)
````

## Scope and Lifetime of Variables
- When we define a function with variables, then those variablesâ€™ scope is limited to that function. In Python, the scope of a variable is an area where a variable is declared. It is called the variableâ€™s local scope.

- We cannot access the local variables from outside of the function. Because the scope is local, those variables are not visible from the outside of the function.

- Note: The inner function does have access to the outer functionâ€™s local scope.

- When we are executing a function, the life of the variables is up to running time. Once we return from the function, those variables get destroyed. So function does no need to remember the value of a variable from its previous call.

## Global Keyword in Function
In Python, global is the keyword used to access the actual global variable from outside the function. 
we use the global keyword for two purposes:

- To declare a global variable inside the function.
- Declaring a variable as global, which makes it available to function to perform the modification.

````python
global_var=5
def f1():
    print("The global variable is:",global_var)
def f2():
    global global_var
    global_var=555
    print("The global variable is:",global_var)
def f3():
    print("The global variable is:",global_var) 
````

````python
f1()
f2()
f3()
````

- In Python, nonlocal is the keyword used to declare a variable that acts as a global variable for a inner function (i.e., function within another function).
- We can use a nonlocal keyword when we want to declare a variable in the local scope but act as a global scope.

````python
def outer_func():
    x=755
    def inner_func():
        nonlocal x
        x=700
        print("The inner x is:",x)
    inner_func()
    print("The outer function x is:",x)
outer_func()
````

# `map()` function in Python
- In Python, the map() function is used to apply some functionality for every element present in the given sequence and generate a new series with a required modification.

````python
s=[1,2,3,4,5,6]
a=map(lambda x:x**2,s)
print(list(a))
````

````python
from functools import reduce
list1 = [20, 13, 4, 8, 9]
mul=reduce(lambda x,y:x*y,list1)
print(mul)
````
