```python
#Write a program to find words with both alphabets and numbers from an input string
str1 = "Emma25 is Data scientist50 and AI Expert"
print("The original string is : " + str1)
res=[]
temp=str1.split()
print(temp)
for i in temp:
    if any(c.isalpha() for c in i) and any(c.isdigit() for c in i):
        res.append(i)
print(res)
```

```python
#Write a function to flatten a list of lists into a single, non-nested list. (e.g., [[1, 2], [3, 4]] becomes [1, 2, 3, 4]).
#not using list comprehension
list1 = [[1, 2], [3, 4], [5, 6, 7]]

def flattend_list(nestedlist):
    flattend=[]
    for item in nestedlist:
        for subitem in item:
            flattend.append(subitem)
    return flattend
a=flattend_list(list1)
a
            
```

```python
#using list comprehension
list1 = [[1, 2], [3, 4], [5, 6, 7]]
def flattend_list(nestedlist):
    return [subitem for item in nestedlist for subitem in item]
a=flattend_list(list1)
a
```
